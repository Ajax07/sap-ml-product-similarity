```python
from app.services.data_loader import DataLoader
from app.services.preprocessing import (
    ProductPreprocessor
)

loader = DataLoader(
    "../data/marketing_sample_for_amazon_com-amazon_fashion_products__20200201_20200430__30k_data.ldjson"
)

df = loader.load_data()

processor = ProductPreprocessor()

processed_df = processor.preprocess(df)

processed_df[[
    "brand",
    "colour",
    "sales_price",
    "weight",
    "weight_cleaned",
    "combined_text"
]].head()
```

```python
from app.services.text_encoder import (
    TextEncoder
)
from app.services.feature_builder import (
    FeatureBuilder
)

# Build structured features
builder = FeatureBuilder()

structured_features = (
    builder
    .build_structured_features(
        processed_df
    )
)

print(
    "Structured shape:",
    structured_features.shape
)

# Text embeddings
encoder = TextEncoder()

text_embeddings = encoder.encode(
    processed_df[
        "combined_text"
    ].tolist()[:100]
)

print(
    "Text embedding shape:",
    text_embeddings.shape
)
```

```python
import numpy as np
```

```python
text_embeddings = encoder.encode(
    processed_df[
        "combined_text"
    ].tolist()
)

structured_weight = 0.35
text_weight = 0.65

final_embeddings = np.hstack([
    structured_features
    * structured_weight,

    text_embeddings
    * text_weight
])

print(final_embeddings.shape)
```

```python
from app.services.similarity_engine import (
    SimilarityEngine
)

engine = SimilarityEngine(
    processed_df,
    final_embeddings
)

sample_product_id = (
    processed_df[
        "uniq_id"
    ].iloc[0]
)

similar_products = (
    engine.find_similar_products(
        product_id=sample_product_id,
        num_similar=5
    )
)

print(
    "Query Product:",
    sample_product_id
)

print(
    "Similar Products:"
)

similar_products
```
