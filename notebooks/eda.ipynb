# Evaluate our Model

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import (
    cosine_similarity
)


def evaluate_similarity_engine(
    df,
    embeddings,
    n_queries=100,
    top_k=5
):

    category_match_scores = []
    brand_match_scores = []
    price_distance_scores = []
    cosine_scores = []

    sample_indices = np.random.choice(
        len(df),
        n_queries,
        replace=False
    )

    similarity_matrix = cosine_similarity(
        embeddings
    )

    for idx in sample_indices:

        query_row = df.iloc[idx]

        similarities = (
            similarity_matrix[idx]
        )

        top_indices = (
            np.argsort(
                similarities
            )[::-1][1:top_k+1]
        )

        retrieved = df.iloc[
            top_indices
        ]

        # -------------------------
        # Category consistency
        # -------------------------
        query_category = (
            query_row[
                "parent___child_category__all"
            ]
        )

        category_match = (
            retrieved[
                "parent___child_category__all"
            ]
            == query_category
        ).mean()

        category_match_scores.append(
            category_match
        )

        # -------------------------
        # Brand consistency
        # -------------------------
        query_brand = (
            query_row["brand"]
        )

        brand_match = (
            retrieved["brand"]
            == query_brand
        ).mean()

        brand_match_scores.append(
            brand_match
        )

        # -------------------------
        # Price similarity
        # -------------------------
        query_price = (
            query_row[
                "sales_price"
            ]
        )

        price_diff = np.abs(
            retrieved[
                "sales_price"
            ].fillna(
                query_price
            )
            - query_price
        ).mean()

        price_distance_scores.append(
            price_diff
        )

        # -------------------------
        # Cosine similarity
        # -------------------------
        cosine_scores.append(
            similarities[
                top_indices
            ].mean()
        )

    results = {

        "Category Consistency":
            np.mean(
                category_match_scores
            ),

        "Brand Consistency":
            np.mean(
                brand_match_scores
            ),

        "Avg Price Distance":
            np.mean(
                price_distance_scores
            ),

        "Avg Cosine Similarity":
            np.mean(
                cosine_scores
            )
    }

    return pd.DataFrame(
        results.items(),
        columns=[
            "Metric",
            "Value"
        ]
    )
```

```python
text_results = (
    evaluate_similarity_engine(
        processed_df,
        final_embeddings,
        n_queries=100,
        top_k=5
    )
)

text_results
```

```python
multimodal_results = (
    evaluate_similarity_engine(
        subset_df,
        multimodal_embeddings,
        n_queries=50,
        top_k=5
    )
)

multimodal_results
```
