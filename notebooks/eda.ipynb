```python
subset_size = 300

subset_df = (
    processed_df
    .head(subset_size)
    .copy()
)

encoder = ImageEncoder()

image_embeddings = []

for idx, url in enumerate(
    subset_df[
        "image_urls__small"
    ]
):

    emb = (
        encoder.encode_image_url(
            url
        )
    )

    image_embeddings.append(
        emb
    )

    if idx % 25 == 0:
        print(
            f"Processed {idx}"
        )

print(
    "Finished:",
    len(image_embeddings)
)
```

```python
valid_embeddings = sum(
    emb is not None
    for emb in image_embeddings
)

print(
    "Valid embeddings:",
    valid_embeddings
)

print(
    "Coverage:",
    round(
        valid_embeddings
        / len(image_embeddings)
        * 100,
        2
    ),
    "%"
)
```

```python
import numpy as np

# Replace failed downloads
# with zero vectors
image_feature_dim = 512

clean_image_embeddings = []

for emb in image_embeddings:

    if emb is None:

        clean_image_embeddings.append(
            np.zeros(
                image_feature_dim
            )
        )

    else:

        clean_image_embeddings.append(
            emb
        )

image_matrix = np.vstack(
    clean_image_embeddings
)

print(
    image_matrix.shape
)
```

```python
subset_text_embeddings = (
    text_embeddings[
        :subset_size
    ]
)

subset_structured_features = (
    structured_features[
        :subset_size
    ]
)

TEXT_WEIGHT = 0.50
STRUCTURED_WEIGHT = 0.20
IMAGE_WEIGHT = 0.30

multimodal_embeddings = (
    np.hstack([
        subset_text_embeddings
        * TEXT_WEIGHT,

        subset_structured_features
        * STRUCTURED_WEIGHT,

        image_matrix
        * IMAGE_WEIGHT
    ])
)

print(
    multimodal_embeddings.shape
)
```

```python
from sklearn.metrics.pairwise import (
    cosine_similarity
)

query_idx = 0

query_vector = (
    multimodal_embeddings[
        query_idx
    ]
    .reshape(1, -1)
)

similarities = (
    cosine_similarity(
        query_vector,
        multimodal_embeddings
    )[0]
)

top_indices = (
    np.argsort(
        similarities
    )[::-1][1:6]
)

query_product = (
    subset_df.iloc[
        query_idx
    ][[
        "product_name",
        "brand",
        "sales_price"
    ]]
)

query_product
```

```python
subset_df.iloc[
    top_indices
][[
    "product_name",
    "brand",
    "sales_price"
]]
```

```python
from sklearn.preprocessing import normalize

subset_text_embeddings = (
    text_embeddings[:subset_size]
)

subset_structured_features = (
    structured_features[:subset_size]
)

# Normalize each modality
normalized_text = normalize(
    subset_text_embeddings
)

normalized_structured = normalize(
    subset_structured_features
)

normalized_image = normalize(
    image_matrix
)

TEXT_WEIGHT = 0.70
STRUCTURED_WEIGHT = 0.15
IMAGE_WEIGHT = 0.15

multimodal_embeddings = np.hstack([
    normalized_text
    * TEXT_WEIGHT,

    normalized_structured
    * STRUCTURED_WEIGHT,

    normalized_image
    * IMAGE_WEIGHT
])

print(
    multimodal_embeddings.shape
)
```
